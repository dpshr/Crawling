```python
from bs4 import BeautifulSoup as bs
```

```python
search_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_sug.pre&fbm=0&acr=1&acq=%EC%98%A4%EB%A6%AC%ED%83%95&qdt=0&ie=utf8'

#params: ({})         
#  URL  key,value      .
res = req.get(search_url, params={'query':''})
res
```

```python
bs(res.text,'lxml')
```

```python
# bs( , )
# -       
#  :  (HTML)    
#  : Lxml > html.parser > html5lib -->   
html = bs(res.text,'lxml')
```

```python
type(res.text), type(html)
```

```python
# CSS 
#1. : HTML   ex) <div><p></p><div> -> p,div
#2. : .  ex) <p class='msg'> -> .msg
#3. : #  ex) <p id='msg'> -> #msg
#4. : *   
#5. : (A>B), (A B),(A+B), (A~B) 
html.select_one('title')
```

```python
html.select_one('title').text
```

```python
#   Tip)
# - HTML id class   
# -,  id class (  )
html.select_one('a.title')
```

```python
html.select_one('a.title').text
```

```python
#     
# .get()
html.select_one('a.title').get('href')
```

```python
# select_one(css):     HTML  
# select(css):     HTML   
html.select('a.title')
```

```python
aTags = html.select('a.title')

for i in range(len(aTags)):
    print(aTags[i].text)
    print(aTags[i].get('href'))
```

```python
aTags = html.select('a.title')

for a in aTags:
    print(a.text)
    print(a.get('href'))
```
