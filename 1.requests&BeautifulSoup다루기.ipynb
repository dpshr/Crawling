```python
from bs4 import BeautifulSoup as bs
```

```python
search_url = 'https://search.naver.com/search.naver?where=nexearch&sm=top_sug.pre&fbm=0&acr=1&acq=%EC%98%A4%EB%A6%AC%ED%83%95&qdt=0&ie=utf8'

#params: 딕셔너리({}) 형태로 값을 구성하여 서버에 전달할 때 사용하는 매개변수 
# 요청한 URL주소에는 동일한 key,value가 없어야 정상적인 결과를 받을 수 있음.
res = req.get(search_url, params={'query':'돈까스'})
res
```

```python
bs(res.text,'lxml')
```

```python
# bs(변환할 데이터, 변환도구)
# - 웹 문서를 파이썬 자료형으로 변환해주는 기능 
# 변환할 데이터: 웹 문서(HTML)을 파이썬 자료형으로 변화 
# 변환도구 : Lxml > html.parser > html5lib --> 변환속도 차이 
html = bs(res.text,'lxml')
```

```python
type(res.text), type(html)
```

```python
# CSS선택자 
#1. 타입선택자: HTML요소의 이름으로 선택 ex) <div><p></p><div> -> p,div
#2. 클래스선택자: .클래스이름으로 선택 ex) <p class='msg'> -> .msg
#3. 아이디선택자: #아이디이름으로 선택 ex) <p id='msg'> -> #msg
#4. 전체선택자: *로 모든 요소를 선택
#5. 계층선택자: 자식(A>B), 자손(A B),근접후행(A+B), 후행(A~B) 
html.select_one('title')
```

```python
html.select_one('title').text
```

```python
# 데이터 추출 Tip)
# -접근한 HTML요소에 id나 class속성이 있는지 체크 
# -없으면, 부모요소의 id나 class속성을 체크(계층적 구조로 생각)
html.select_one('a.title')
```

```python
html.select_one('a.title').text
```

```python
# 접근한 요소의 속성값 가져오는 방법
# .get(속성이름)
html.select_one('a.title').get('href')
```

```python
# select_one(css선택자): 선택자 조건에 해당하는 하나의 HTML요소를 반환 
# select(css선택자): 선택자 조건에 해당하는 모든 HTML요소를 찾아서 리스트로 반환
html.select('a.title')
```

```python
aTags = html.select('a.title')

for i in range(len(aTags)):
    print(aTags[i].text)
    print(aTags[i].get('href'))
```

```python
aTags = html.select('a.title')

for a in aTags:
    print(a.text)
    print(a.get('href'))
```
