```python
import pandas as pd
import requests as req 
from bs4 import BeautifulSoup as bs

# 서버에서 브라우저가 요청한 것 처럼 인식할 수 있도록 요청헤더에 'user-agent'값을 같이 보내줘야 한다.(★★★)
headerInfo = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'
}
```

```python
melon_url = 'https://www.melon.com/chart/index.htm'
res = req.get(melon_url, headers=headerInfo)
```

```python
res.text
```

```python
html = bs(res.text, 'lxml') # 문자정보를 html로 변환
```

```python
# a 태그 안에있는 콘탠츠를 가져오겠다.
# >: 자식선택자
# ' ' : 자손 선택자 
html.select('div.ellipsis.rank01 > span > a')
```

### 1. 멜론 음원 데이터 수집
- 수집할 데이터 : 순위, 곡명, 가수
- DataFrame으로 만들기
- 엑셀 파일로 저장하기
  **'가수'의 경우, 1곡에 2명 이상이 참여한 곳도 있으므로, 하나로 묶어서 가져오는 방법에 대한 고민 필요

```python
melon_rank =html.select('div.wrap.t_center>span.rank')
melon_rank
```

```python
melon_title = html.select('div.ellipsis.rank01>span')
melon_title
```

```python
melon_singer = html.select('span.checkEllipsis')
melon_singer
```

```python
len(melon_singer)
```

```python
for i in range(len(melon_title)):
    print(melon_rank[i+1].text.strip(),end=' ')
    print(melon_singer[i].text.strip(),end=' ')
    print(melon_title[i].text.strip())
```

```python
list_mr = []
list_sg = []
list_tl = []
for i in range(len(melon_title)):
    list_mr.append(melon_rank[i+1].text)
    list_sg.append(melon_singer[i].text)
    list_tl.append(melon_title[i].text.strip())
```

```python
data= {"순위" : list_mr,"가수" : list_sg,"제목" : list_tl}
df = pd.DataFrame(data)
df.set_index("순위",inplace=True)
df
```

```python
# 2. csv파일로 내보내기 
df.to_csv('멜론차트.csv' , encoding = 'utf-8')
```
